```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} は大人です")
        else:
            print(f"{self.name} は大人ではありません")
                
members = [
    Human("太郎",20),
    Human("花子",25),
    Human("陽介",35),
    Human("トム",12)
]
    
for member in members:
    member.check_adult()
```
